List of Columns with Null values

```python
null_bool_list = list()
for i in df.columns:
    if (df[i] == " ").sum():
        null_bool_list.append(i)

print(null_bool_list)
```

Optimization of Data Types

```python
category_convert = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'DeviceProtection', 'OnlineBackup', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges', 'Churn']

df["TotalCharges"] = df["TotalCharges"].replace({" ": 0})

for col in category_convert:
    df[col] = df[col].astype('category')
df['tenure'] = df['tenure'].astype('int8')
df['MonthlyCharges'] = df['MonthlyCharges'].astype('float16')
df['TotalCharges'] = df['TotalCharges'].astype('float32')
```

Simplifying Multi-class Categoricals

```python
no_internet_service = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
for col in no_internet_service:
    df[col] = df[col].replace({"No internet service": "No"})
df['MultipleLines'] = df['MultipleLines'].replace({"No phone service": "No"});
```

Binary Encoding

```python
encoding_cols = ['SeniorCitizen']
df[encoding_cols] = df[encoding_cols].replace({0: "No", 1: "Yes"});
```

Feature Creation

```python
service_cols = ['PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

df['TotalServices'] = (df[service_cols] == 'Yes').sum(axis=1)
df['TotalServices'] = df['TotalServices'].astype('Int8')
```
